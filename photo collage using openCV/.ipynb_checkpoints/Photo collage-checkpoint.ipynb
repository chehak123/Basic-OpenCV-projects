```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread('img/sansa.jpg')
img1=cv2.resize(img1,(500,400))
img2=cv2.imread('img/jon.jpg')
img2=cv2.resize(img2,(500,400))
img3=cv2.imread('img/daenerys.jpg')
img3=cv2.resize(img3,(500,400))
img4=cv2.imread('img/tyrion.jpg')
img4=cv2.resize(img4,(500,400))

# img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
# plt.imshow(img4)
# plt.show()
```

```python

border1=np.zeros((30,1030,3), dtype='uint8')
border1[ : , : , : ]= 0

border2=np.zeros((400,30,3), dtype='uint8')
border2[ : , : , : ]= 0

got1=np.hstack((img1,border2,img3))
got2=np.hstack((img2,border2,img4))

gotfinal=np.vstack((got1,border1,got2))
plt.imshow(gotfinal)
cv2.imshow('img',gotfinal)
cv2.waitKey(0)
plt.show()
print(got1.shape)
```
